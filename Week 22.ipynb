```r
library(tidyverse)
library(repr)
library(ggdark)
```

```r
records <- read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-05-25/records.csv')
drivers <- read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-05-25/drivers.csv')
```

```r
records %>% head()
```

```r
options(repr.plot.res = 260, repr.plot.width = 10, repr.plot.height = 7)

dark_line_colours = c("#ff6363", "#ffbd69")

records %>%
    filter(track == "Rainbow Road") %>%
    ggplot(aes(date, time, color = type, shape = shortcut)) + 
    geom_line() + 
    geom_point() + 
    dark_theme_minimal() + 
    labs(title = "Rainbow Road Records over Time", 
         shape = "Shortcut Used", 
         color = "Track Type",
         y = "Record\nTime (s)",
         x = "Record Date",
         caption = "Tidy Tuesday Week 22\nMario Kart 64 World Records\nCreated by: Michal Jagodzinski") + 
    theme(plot.title.position = "plot",
          axis.title.y = element_text(angle = 0, vjust = 0.5, margin = margin(0.5, 0.5, 0.5, 0.5, "cm")),
          axis.title.x = element_text(margin = margin(0.5, 0.5, 0.5, 0.5, "cm")),
          panel.grid.minor = element_line(color = "#543864", linetype = "dashed"),
          panel.grid.major = element_line(color = "#543864", linetype = "dashed"),
          plot.background = element_rect(fill = "#202040")) + 
    scale_color_manual(values = dark_line_colours)
```
